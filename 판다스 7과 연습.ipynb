## 인덱스 리셋하기
- 데이터프레임을 붙이고 나서 인덱스를 다시 재정렬

```python
import pandas as pd
row_concat_i = pd.concat([df1, df2, df3, df3], ignore_index = True)
row_concat_i
```

## concat 함수
- axis = 0 / 'index' (기본값) - 행방향
- axis = 1 / 'columns' - 열방향
- 인덱스 재정렬 : ignore_index = True

```python
col_concat = pd.concat([df1, df2, df3], axis = 'columns')
col_concat
```

```python
col_concat['A'] # 행 이름으로 추출하면 세 열이 나옴
```

```python
col_concat['new_col_list'] = ['n1', 'n2', 'n3', 'n4']
col_concat
```

```python
col_concat['new_col_series'] = pd.Series(['n1', 'n2', 'n3', 'n4'])
col_concat
```

```python
col_concat2 = pd.concat([df1, df2, df3], axis = 'columns', ignore_index = True)
```

```python
col_concat2
```

## concat 함수 join 옵션 사용하여 연결
- inner, outer 옵션만 존재
- 데이터프레임으로 출력됨

```python
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
```

```python
print(pd.concat([df1, df2, df3], join = 'inner'))
```

```python
test = pd.concat([df1, df3], ignore_index = False, join = 'inner')
```
